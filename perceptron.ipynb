```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['label'] = iris.target
```

```python
data = np.array(df.iloc[:100, [0, 1, -1]])
X, y = data[:,:-1], data[:,-1]
y = np.array([1 if i == 1 else -1 for i in y])
```

```python
class Perceptron:
    def __init__(self,learning_rate=0.01,feature_dim=2,thrs=0):
        self.learning_rate = learning_rate
        self.W = np.random.normal(size=(feature_dim,1))
        self.b = 0
        self.thrs = thrs

    def train(self,X,y):
        total_error = 1
        X = np.array(X)
        y = np.array(y.reshape(-1,1))
        d_W_pre_calculated = - X*y
        d_b_pre_calculated = - y
        while total_error>self.thrs:
            prediction = np.matmul(X, self.W)+self.b
            error = np.sign(prediction)!=y
            total_error = np.sum(error)
            print(total_error/len(X))
            d_W = np.sum(error*d_W_pre_calculated,axis = 0)
            d_b = np.sum(error*d_b_pre_calculated)
            self.W -= self.learning_rate * d_W.reshape(-1,1)
            self.b -= self.learning_rate * d_b
```

```python
perceptron = Perceptron(1,len(X[0]))
perceptron.train(X, y)
```

```python
class PerceptronDual:
    def __init__(self,learning_rate=0.01,data_size=50,thrs=0):
        self.learning_rate = learning_rate
        self.alpha = np.zeros((data_size,1))
        self.b = 0
        self.thrs = thrs
    
    def train(self,X,y):
        total_error = 1
        X = np.array(X)
        y = np.array(y.reshape(-1,1))
        Xy = X*y
        G = np.matmul(Xy,Xy.T)
        print(G.shape)
        while total_error>self.thrs:
            error = (np.matmul(G,self.alpha)+y*self.b)<=0
            total_error = np.sum(error)
            print(total_error/len(X))
            self.alpha += error*self.learning_rate
            self.b += self.learning_rate*np.sum(error*y)
```

```python
perceptron = PerceptronDual(1,len(X))
perceptron.train(X, y)
```
